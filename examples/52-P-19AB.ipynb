Performing basic calculations of centrifugal pump.

```python
# Importing the libray
import pump
Q_ = pump.Q_

fluid = pump.Fluid(
    name="Hidrocarboneto Condensado",
    density=Q_(763.02, "kg/m**3")
)
fluid
```

```python
rated_flow = Q_(18.51, "m**3/h")
design_point = pump.DesignPoint(
    fluid=fluid,
    capacity=rated_flow,
    differential_head=Q_(28.53, "m"),
    inlet_pressure=Q_(2.01, "kgf/cm**2")
)
design_point
```

```python
design_point.outlet_pressure
```

```python
design_point.specific_energy
```

```python
design_point.power_output
```
