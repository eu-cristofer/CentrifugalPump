Performing basic calculations of a centrifugal pump.

## Creating an Fluid instance

```python
# Importing the libray
import pump
Q_ = pump.Q_

fluid = pump.Fluid(
    "Acid water",
    Q_(994.86, "kg/m**3"),
    viscosity=Q_(1, "cP")
)
print(fluid)
fluid
```

## Creating a BasePoint instance

`BasePoint` is class developed to implement the commom foundations of all classes applied to handle Centrifugal Pumps Performance Point .

```python
rated_flow = Q_(23.44, "m**3/h")
base_point = pump.BasePoint(fluid, rated_flow)
print(base_point)
base_point
```

```python
normal_flow =  Q_(19.53, "m**3/h")
rated_head = Q_(133.09, "m")

# Instantiating with custom proterties
base_point_2 = pump.BasePoint(fluid, normal_flow, differential_head=rated_head)
print(base_point_2)
base_point_2
```

## Creating a Design Point.

The design point is basically a conteiner to hold rated point data

```python
# Suction Pressure Condition
rated_point = pump.DesignPoint(
    fluid,
    rated_flow,
    rated_head,
    inlet_pressure=Q_(1.81, "kgf/cm**2")
)
print(rated_point)
rated_point
```

```python
# Functionalites sample
rated_point.outlet_pressure
```

```python
rated_point.specific_energy
```

```python
# Design Pressure Condition and API consideration +/- 8 % @ shutoff
shutoff_point = pump.DesignPoint(
    fluid,
    rated_flow,
    differential_head=Q_(161 * 1.08, "m"),
    inlet_pressure=Q_(3.88, "kgf/cm**2")
)
shutoff_point.outlet_pressure
```
